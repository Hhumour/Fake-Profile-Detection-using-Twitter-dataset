```python
import sys
import datetime
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import gender_guesser.detector as gender
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
```

```python
genuine_users = pd.read_csv("data/users.csv")
fake_users = pd.read_csv("data/fusers.csv")
# print genuine_users.columns
# print genuine_users.describe()
#print fake_users.describe()
x=pd.concat([genuine_users,fake_users])   
y=len(fake_users)*[0] + len(genuine_users)*[1]
x.dtypes
```

```python
def predict_sex(name):
    sex_predictor=gender.Detector(case_sensitive=False)
    first_name=name.str.split(' ').str.get(0)
    sex=first_name.apply(sex_predictor.get_gender)
    sex_dict={'female':-2,'mostly_female':-1,'unknown':0,'mostly_male':1,'male':2,'andy':3}
    sex_code=sex.map(sex_dict).astype(int)
    return(sex_code)
```

```python
lang_list = list(enumerate(np.unique(x['lang']))) #it assigns an index starting from 0 to unique values of lang column
lang_dict = { name : i for i, name in lang_list } #list is converted to dictionary use type() to check
x.loc[:,'lang_code'] = x['lang'].map( lambda x: lang_dict[x]).astype(int) # it creates a new column lang_code and map the values
sex_predictor = gender.Detector(case_sensitive=False)
x.loc[:,'sex_code']=predict_sex(x['name'])
feature_columns_to_use = ['statuses_count','followers_count','friends_count','favourites_count','listed_count','sex_code','lang_code']
x=x.loc[:,feature_columns_to_use]
x
```

```python
X_train,X_test,y_train,y_test = train_test_split(x, y, test_size=0.25, random_state=44)
```

```python
def Using_random_Classifier(X_train,y_train,X_test):
    clf1=RandomForestClassifier(n_estimators=100)
    clf1.fit(X_train,y_train)
    y_pred=clf1.predict(X_test)
    return(accuracy_score(y_test, y_pred))
    
```

```python
def Using_Support_Vector(X_train,y_train,X_test):
    clf2=SVC()
    clf2.fit(X_train,y_train)
    y_pred=clf2.predict(X_test)
    return(accuracy_score(y_test, y_pred))
    
```

```python
RC_score=Using_random_Classifier(X_train,y_train,X_test)
SVM_score=Using_Support_Vector(X_train,y_train,X_test)
```
